<h1>Question Set A

```python
#1) Create a Dictionary with Roll Number as Key and Name as Value
# Create dictionary for 5 students
students = {}
for i in range(1, 6):
    roll = input(f"Enter roll number for student {i}: ")
    name = input(f"Enter name for student {i}: ")
    students[roll] = name

# Display all key-value pairs
print("\nStudent Dictionary:")
for roll, name in students.items():
    print(f"Roll No: {roll}, Name: {name}")
```

```python
#2) Count Frequency of Each Word in a Sentence
# Input sentence
sentence = input("Enter a sentence: ")

# Split into words and count frequency
words = sentence.split()
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1  # get() returns 0 if word not present

# Display frequency
print("\nWord Frequency:", freq)
```

```python
#### 3) Merge Two Dictionaries Entered by User
# Input first dictionary
dict1 = {}
n1 = int(input("How many items in first dictionary? "))
for _ in range(n1):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

# Input second dictionary
dict2 = {}
n2 = int(input("How many items in second dictionary? "))
for _ in range(n2):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value

# Merge dictionaries
merged_dict = {**dict1, **dict2}  # Python 3.5+ way

print("\nMerged Dictionary:", merged_dict)
```

```python
#4) Find Key with Maximum Value
# Example dictionary
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'David': 95}

# Find key with max value
max_key = max(scores, key=scores.get)
print("Key with maximum value:", max_key)
print("Maximum value:", scores[max_key])
```

```python
#5) Check if a Key Exists in a Dictionary
# Example dictionary
students = {'101': 'Alice', '102': 'Bob', '103': 'Charlie'}

# Input key to check
key = input("Enter key to check: ")
if key in students:
    print(f"Key '{key}' exists in the dictionary.")
else:
    print(f"Key '{key}' does not exist in the dictionary.")
```

<h1>Question Set B

```python
#1) Remove a Key from a Dictionary and Handle Non-Existent Key
# Example dictionary
students = {'101': 'Alice', '102': 'Bob', '103': 'Charlie'}

# Input key to remove
key = input("Enter key to remove: ")

# Remove key safely
if key in students:
    students.pop(key)
    print(f"Key '{key}' removed. Updated dictionary:", students)
else:
    print(f"Key '{key}' does not exist. Dictionary remains:", students)
```

```python
#2) Sort a Dictionary by Values in Descending Order
# Example dictionary
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'David': 95}

# Sort by values in descending order
sorted_dict = dict(sorted(scores.items(), key=lambda item: item[1], reverse=True))

print("Dictionary sorted by values (descending):", sorted_dict)
```

```python
#3) Create a Dictionary from Two Lists
# Input lists
keys = input("Enter keys separated by space: ").split()
values = input("Enter values separated by space: ").split()

# Create dictionary
my_dict = dict(zip(keys, values))

print("Dictionary created from two lists:", my_dict)
```

<h1>Question Set C

```python
#1) Find All Keys with the Same Value in a Dictionary
# Example dictionary
my_dict = {'a': 10, 'b': 20, 'c': 10, 'd': 30, 'e': 20}

# Input value to search for
value_to_find = int(input("Enter value to find keys for: "))

# Find keys with that value
keys_with_value = [key for key, value in my_dict.items() if value == value_to_find]

# Display results
if keys_with_value:
    print(f"Keys with value {value_to_find}:", keys_with_value)
else:
    print(f"No keys found with value {value_to_find}")
```

```python
#2) Create a Dictionary Where Keys are Numbers from 1 to n and Values are Their Cubes
# Input n
n = int(input("Enter a number n: "))

# Create dictionary with cubes
cube_dict = {i: i**3 for i in range(1, n+1)}

print("Dictionary of cubes:", cube_dict)
```
