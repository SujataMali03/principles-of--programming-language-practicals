<h1>Question Set A

```python
#1) Write a program to count the number of vowels in a given string.
# Program to count vowels in a string

text = input("Enter a string: ")
count = 0

for ch in text:
    if ch.lower() in 'aeiou':
        count += 1

print("Number of vowels:", count)
```

```python
#2) Write a program to check if a string is a palindrome (reads the same forwards and
#backwards).
#Example: "madam" is a palindrome, "python" is not.
# Program to check palindrome string

text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
#3) Write a program to reverse a string without using built-in reverse functions.
# Program to reverse a string without built-in functions

text = input("Enter a string: ")
reverse = ""

for ch in text:
    reverse = ch + reverse

print("Reversed string:", reverse)
```

```python
#4) Write a program to count the frequency of each character in a string.
#Example: For "hello", output should be h:1, e:1, l:2, o:1
# Program to count frequency of each character in a string

text = input("Enter a string: ")
freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

for key in freq:
    print(key + ":", freq[key])
```

```python
#5) Write a program to remove all whitespace from a string.
# Program to remove all whitespace from a string

text = input("Enter a string: ")
result = ""

for ch in text:
    if ch != " ":
        result += ch

print("String without whitespace:", result)
```

```python
#2) Write a program to check if two strings are anagrams (contain same characters in different
#order).
# Program to check if two strings are anagrams

str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Remove spaces and convert to lowercase for uniformity
str1 = str1.replace(" ", "").lower()
str2 = str2.replace(" ", "").lower()

# Check if sorted characters of both strings are the same
if sorted(str1) == sorted(str2):
    print("Anagrams")
else:
    print("Not anagrams")
```

```python
#3) Write a program to convert a string to title case without using the title() method.
#: "python programming" should become "Python Programming"
# Program to convert a string to title case without using title()

text = input("Enter a string: ")
words = text.split()  # Split string into words
title_case = ""

for word in words:
    title_case += word[0].upper() + word[1:].lower() + " "

# Remove trailing space
title_case = title_case.strip()

print("Title Case:", title_case)
```

```python
#4) Write a program to replace all occurrences of a substring with another substring in a given
#string.
# Program to replace all occurrences of a substring

text = input("Enter the original string: ")
old_sub = input("Enter the substring to replace: ")
new_sub = input("Enter the new substring: ")

# Replace all occurrences
new_text = text.replace(old_sub, new_sub)

print("Updated string:", new_text)
```

```python
#1) Write a program to find the longest word in a sentence.
# Program to find the longest word in a sentence

sentence = input("Enter a sentence: ")
words = sentence.split()  # Split sentence into words

longest_word = words[0]  # Assume first word is longest

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print("Longest word:", longest_word)
```

```python
#2) Write a program to check if a string contains only digits, only alphabets, or both.
# Program to check type of characters in a string

text = input("Enter a string: ")

if text.isdigit():
    print("The string contains only digits.")
elif text.isalpha():
    print("The string contains only alphabets.")
else:
    print("The string contains both digits and alphabets or other characters.")
```
