<h1>Question Set A

```python
#1) Function to find the maximum of three numbers
def max_of_three(a, b, c):
    return max(a, b, c)

# Example
print("Maximum number is:", max_of_three(10, 25, 15))
```

```python
#2) Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Example
num = 17
print(num, "is prime?" , is_prime(num))
```

```python
#3) Function to calculate factorial (iterative and recursive)
# Iterative
def factorial_iterative(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

# Recursive
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)

# Example
num = 5
print("Iterative factorial of", num, "is", factorial_iterative(num))
print("Recursive factorial of", num, "is", factorial_recursive(num))
```

```python
#4) Function to return unique elements from a list
def unique_elements(lst):
    return list(set(lst))

# Example
my_list = [1, 2, 2, 3, 4, 4, 5]
print("Unique elements:", unique_elements(my_list))
```

```python
#5) Function to calculate the sum of all even numbers in a list
def sum_even_numbers(lst):
    return sum(x for x in lst if x % 2 == 0)

# Example
my_list = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_even_numbers(my_list))
```

```python
#1) Function to convert temperature
def convert_temperature(value, unit):
    """
    Converts temperature between Celsius and Fahrenheit.
    
    :param value: numeric temperature value
    :param unit: 'C' for Celsius, 'F' for Fahrenheit
    :return: converted temperature
    """
    if unit.upper() == 'C':
        # Celsius to Fahrenheit
        return (value * 9/5) + 32
    elif unit.upper() == 'F':
        # Fahrenheit to Celsius
        return (value - 32) * 5/9
    else:
        return "Invalid unit! Use 'C' or 'F'."

# Example usage:
print(convert_temperature(0, 'C'))   
print(convert_temperature(32, 'F'))
```

```python
#2) Function to count vowels and consonants in a string
def count_vowels_consonants(text):
    vowels = 'aeiouAEIOU'
    text = text.replace(" ", "")  # remove spaces
    vowel_count = sum(1 for char in text if char in vowels)
    consonant_count = sum(1 for char in text if char.isalpha() and char not in vowels)
    return vowel_count, consonant_count

# Example usage:
v, c = count_vowels_consonants("Hello World")
print(f"Vowels: {v}, Consonants: {c}")  
```

```python
#3) Function to calculate average of variable number of arguments
def average(*args):
    if len(args) == 0:
        return 0
    return sum(args) / len(args)

# Example usage:
print(average(2, 4, 6))     
print(average(10, 20, 30, 40)) 
```

```python
#4) Lambda function to find maximum of three numbers
max_of_three = lambda a, b, c: max(a, b, c)

# Example usage:
print(max_of_three(5, 10, 3))  
```

```python
#1) Recursive function to calculate the sum of digits of a number
def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_of_digits(n // 10)

# Example usage:
number = 12345
result = sum_of_digits(number)
print(f"Sum of digits of {number} is {result}")
```

```python
#2) Function to count even and odd numbers in a list
def count_even_odd(numbers):
    result = {'even': 0, 'odd': 0}
    for num in numbers:
        if num % 2 == 0:
            result['even'] += 1
        else:
            result['odd'] += 1
    return result

# Example usage:
numbers_list = [1, 2, 3, 4, 5]
result_dict = count_even_odd(numbers_list)
print(f"Count of even and odd numbers: {result_dict}")
```
