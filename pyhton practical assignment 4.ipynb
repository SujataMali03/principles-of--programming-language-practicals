<h1>Question Set A

```python
#1) Union, Intersection, and Difference of Two Sets
# Input two sets from user
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))

# Union
union_set = set1.union(set2)
# Intersection
intersection_set = set1.intersection(set2)
# Difference (set1 - set2)
difference_set = set1.difference(set2)

print("\nUnion:", union_set)
print("Intersection:", intersection_set)
print("Difference (set1 - set2):", difference_set)
```

```python
#2) Check if a Set is a Subset of Another Set
# Input two sets
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))

# Check subset
if set1.issubset(set2):
    print("First set is a subset of the second set.")
else:
    print("First set is NOT a subset of the second set.")
```

```python
#3) Remove Duplicates from a List Using Sets
# Input list with possible duplicates
lst = list(map(int, input("Enter elements of list separated by space: ").split()))

# Remove duplicates
unique_elements = list(set(lst))

print("List after removing duplicates:", unique_elements)
```

```python
#4) Find Elements Present in First Set but Not in Second Set
# Input two sets
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))

# Elements in set1 but not in set2
difference_set = set1 - set2

print("Elements in first set but not in second set:", difference_set)
```

```python
#5) Check if Two Sets are Disjoint
# Input two sets
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))

# Check disjoint
if set1.isdisjoint(set2):
    print("The two sets are disjoint (no common elements).")
else:
    print("The two sets are not disjoint (they have common elements).")
```

<h1>Question Set B

```python
#1) Program to find symmetric difference between two sets
# Symmetric difference between two sets

# Input sets from user
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))

# Symmetric difference
sym_diff = set1.symmetric_difference(set2)

print("Symmetric difference of the two sets is:", sym_diff)
```

```python
#2) Program to add multiple elements to a set at once from user input
# Adding multiple elements to a set

my_set = set()  # Empty set

# Input elements separated by space
elements = input("Enter elements to add to the set separated by space: ").split()

# Adding elements to set
my_set.update(elements)

print("Set after adding elements:", my_set)
```

```python
#3) Program to find the length of a set and check if a specific element exists
# Set length and element existence check

my_set = {10, 20, 30, 40, 50}

# Find length
length = len(my_set)
print("Length of the set:", length)

# Check if element exists
element = int(input("Enter an element to check: "))
if element in my_set:
    print(element, "exists in the set.")
else:
    print(element, "does not exist in the set.")
```

<h1>Question Set C

```python
#1) Program to compare two sentences
# Input two sentences
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# Convert sentences to sets of words
set1 = set(sentence1.lower().split())
set2 = set(sentence2.lower().split())

# Common words
common_words = set1.intersection(set2)

# Words unique to each sentence
unique_to_first = set1 - set2
unique_to_second = set2 - set1

# Output
print("Common words:", common_words)
print("Words unique to first sentence:", unique_to_first)
print("Words unique to second sentence:", unique_to_second)
```

```python
#2) Program to perform all set operations on three sets
# Input three sets from user
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))
set3 = set(map(int, input("Enter elements of third set separated by space: ").split()))

# Union
union_all = set1.union(set2, set3)

# Intersection
intersection_all = set1.intersection(set2, set3)

# Differences
diff_1_2 = set1 - set2
diff_1_3 = set1 - set3
diff_2_1 = set2 - set1
diff_2_3 = set2 - set3
diff_3_1 = set3 - set1
diff_3_2 = set3 - set2

# Symmetric differences
sym_diff_1_2 = set1.symmetric_difference(set2)
sym_diff_all = (set1.symmetric_difference(set2)).symmetric_difference(set3)

# Output
print("Union of three sets:", union_all)
print("Intersection of three sets:", intersection_all)
print("Difference set1 - set2:", diff_1_2)
print("Difference set1 - set3:", diff_1_3)
print("Difference set2 - set1:", diff_2_1)
print("Difference set2 - set3:", diff_2_3)
print("Difference set3 - set1:", diff_3_1)
print("Difference set3 - set2:", diff_3_2)
print("Symmetric difference between set1 and set2:", sym_diff_1_2)
print("Symmetric difference among all three sets:", sym_diff_all)
```
